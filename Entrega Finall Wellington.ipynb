```python
# input das informações na variávil "info"
info = response.json()
# pprint.pprint(info)

# armazenando as informações do campo Value na variáel "tab"
tab = pd.DataFrame(info["value"])

print(tab)
```

```python
# tratamento dos campos

tab["Valor"] = tab["Valor"].map("{:,.2f}".format)

tab["Quantidade"] = tab["Quantidade"].map("{:,.0f}".format)

tab["Data"] = pd.to_datetime(tab.Data)
tab["Data"] = tab["Data"].dt.strftime('%d/%m/%Y')

tab
```

```python
# input das informações na variável "info3"
info3 = response3.json()

# armazenando as informações do campo Value na variáel "tab3"
tab3 = pd.DataFrame(info3["value"])

print(tab3)
```

```python
#verificando se existe algum campo nulo na tabela

print(tab3.isnull().sum())

tab3.info()
```

```python
# Renomeando colunas

tab3 = tab3.rename(columns={'DataReferencia': 'Data Ref', 'DesvioPadrao' : 'Desvio Padrao', 'numeroRespondentes': 'Qtde Respondentes', 'baseCalculo': 'Base Calc.'})

tab3.info()
```

# Exportando para o banco de Dados

```python
## Exportando a tabela produtos como um banco de dados

# Conectando ao banco de dados
entrega = sqlite3.connect('wellington.db')

#Escrevendo DataFrame tab na tabela api1
tab.to_sql('api1', entrega, if_exists='replace', index=False)

#Escrevendo DataFrame tab2 na tabela api2
tab2.to_sql('api2', entrega, if_exists='replace', index=False)

#escrevendo Dataframe tab3 na tabela api3
tab3.to_sql('api3', entrega, if_exists='replace', index=False)

# Fechando a conexão com o banco de dados
entrega.close()

```
